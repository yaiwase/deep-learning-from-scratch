```python
import numpy as np
```

```python
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import math
```

```python
# データの作成
x = np.arange(0, (3.14)*2, 0.025) 
y = np.sin(x)

# グラフの描画
plt.plot(x, y)
plt.grid(which='major',color='black',linestyle=':')
plt.grid(which='minor',color='black',linestyle=':')
plt.show()
```

```python
x = np.arange(0, (3.14)*2, 0.025) 
y = np.cos(x)

plt.plot(x, y)
plt.grid(which='major',color='black',linestyle=':')
plt.grid(which='minor',color='black',linestyle=':')
plt.show()
```

```python
x = np.arange((-3.14)/2, (3.14)/2, 0.00001) 
y = np.tan(x)

plt.axis([(-3.14)/2, (3.14)/2, (-3.14)*2, (3.14)*2])
plt.plot(x,y)
plt.grid(which='major',color='black',linestyle=':')
plt.grid(which='minor',color='black',linestyle=':')
plt.show()
```

```python
x = np.arange(0, (3.14)*2, 0.025) 
y = np.exp(x)

plt.plot(x,y)
plt.grid(which='major',color='black',linestyle=':')
plt.grid(which='minor',color='black',linestyle=':')
plt.show()
```

```python
x = np.arange(0.001, (3.14*2), 0.025) 
y = np.log(x)

plt.plot(x,y)
plt.grid(which='major',color='darkgreen',linestyle='dashed',linewidth = 1.0)
plt.grid(which='minor',color='darkgreen',linestyle='dashed',linewidth = 1.0)

plt.show()
```

```python
x = np.arange(-1.0, 1.0, 0.00001) 
y = np.arcsin(x)

plt.plot(x,y)
plt.grid(which='major',color='darkgreen',linestyle='dashed',linewidth = 1.0)
plt.grid(which='minor',color='darkgreen',linestyle='dashed',linewidth = 1.0)

plt.show()
```

```python
x = np.arange(-1, 1, 0.00001) 
y = np.arccos(x)

plt.plot(x,y)
plt.grid(which='major',color='darkgreen',linestyle='dashed',linewidth = 1.0)
plt.grid(which='minor',color='darkgreen',linestyle='dashed',linewidth = 1.0)

plt.show()
```

```python
x = np.arange((-3.14)*2, (3.14)*2, 0.025) 
y = np.arctan(x)

plt.plot(x,y)
plt.grid(which='major',color='darkgreen',linestyle='dashed',linewidth = 1.0)
plt.grid(which='minor',color='darkgreen',linestyle='dashed',linewidth = 1.0)

plt.show()
```

```python
# normal distribution
mu = 0
variance = 1
sigma = math.sqrt(variance)

x = np.linspace( mu-(4*variance), mu+(4*variance), 100)
plt.plot(x,mlab.normpdf(x, mu, sigma))

plt.grid(which='major',color='darkgreen',linestyle=':',linewidth = 1.0)
plt.grid(which='minor',color='darkgreen',linestyle=':',linewidth = 1.0)
plt.show()
```

```python
x = np.arange(0, (3.14)*2, 0.025) 
y1 = np.sin(x)
y2 = np.cos(x)

plt.title("sin & cos")
plt.plot(x, y1,label="sin")
plt.plot(x, y2,label="cos",linestyle='--')
plt.grid(which='major',color='darkgreen',linestyle=':')
plt.grid(which='minor',color='darkgreen',linestyle=':')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
